```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline   #grafikleri görebilmek için
```

```python
x=["1.ABD","2.ÇİN","3.JAPONYA","4.ALMANYA","5.İNGİLTERE","6.FRANSA"]
y=[20410,14090,5170,4210,2940,2930]   #Ülke geliri
z=[63039,9957,40649,51158,44162,44915]  #Kişi başına düşen milli gelir
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,z)
```

```python
plt.plot(x,y,"r")
```

```python
plt.plot(x,y,"r")
plt.xlabel("Ülkeler")
plt.ylabel("Milli Gelir")
plt.title("Ülkelere Göre Milli Gelir")
plt.show() #başka idelerde gösterebilmek için
```

### Subplot

```python
plt.subplot(1,2,1)  # 1 satırlı 2 sutunlu bir grafik oluştur
plt.plot(x,y,"r--")
plt.subplot(1,2,2)
plt.plot(x,z,"y*-") #2. grafiğim sarı olsun
plt.tight_layout()  #grafikler iç içe girmesin diye. düzeltebildiği kadar düzeltiyor
```

### Figure

```python
f=plt.figure()
axes=f.add_axes([0.1,0.1,0.9,0.9])  # boş bir dikdörtgen oluşturmak için
axes.plot(x,y)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
plt.show()
```

### İç içe 2 grafik

```python
f=plt.figure()
axes1=f.add_axes([0.1,0.1,0.9,0.9])
axes2=f.add_axes([0.5,0.5,0.4,0.4])
axes1.plot(x,y,"y")
axes2.plot(x,z,"r")
plt.show()
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2)  #içi boş 1 satır 2 sutun şeklinde dörtgen verdi

axes[0].plot(x,y,"b")    #ilk satır ilk sutun -> 0
axes[0].set_title("MİLLİ GELİR")

axes[1].plot(x,z,"y")     #ilk satır 2. sutun --> 1
axes[1].set_title("KİŞİ BAŞINA")

plt.tight_layout()
```

### Renkler ve Çizgiler

linewidth = çizginin kalınlığı
linestyle = çizginin tipi : - , /-.  gibi
marker : her değer için belirteç o,+ gibi
marker boyutu = marker'ın boyutu
markerfacecolor= marker'ın içerisindeki renk
markeredgewidth=marker çevresinin boyutu
alpha=saydamlık değeri

```python
f=plt.figure()
axes=f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,color="red",linewidth=1,linestyle="-",marker="o",markersize=20,markerfacecolor="yellow",markeredgewidth=4,markeredgecolor="blue",alpha=0.9)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
f=plt.figure()
axes=f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,color="red",linewidth=1,linestyle="-")
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")

axes.set_xlim([2,6])   #2.değerden 6. değere kadar. 2 dahil değil
axes.set_ylim([2000,7500])
```

### Grafik Boyutu ve Resim Olarak Kaydetme

```python
f=plt.figure(figsize=(10,6))  #figsize ile grafiğin boyutunu değiştirmek, ilk değer x eksninde büyütüyori ikinci değer y eksinide büyütüyor
axes=f.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2,figsize=(13,5))  #Ülke adları birbirine girmişti, ayırdık

axes[0].plot(x,y,"b")
axes[0].set_title("MİLLİ GELİR")

axes[1].plot(x,z,"y")
axes[1].set_title("KİŞİ BAŞINA")

plt.tight_layout()
```

```python
fig.savefig("resim.png")  #figure'i resim olarak kaydetti.
```

```python
fig=plt.figure(figsize=(10,5))

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y,"r",label="milli gelir")  #öncelikle çizgilerin ne olduğuna dair label ekliyor
ax.plot(x,z,"y",label="kişi başına")

ax.legend(loc=5) #legend --> çigilerin bilgisini gösteren kutucuk, location değiştirebilir.
```

### Pie Chart

```python
labels = 'TÜRKİYE', 'ABD', 'ÇİN', 'ALMANYA'
sizes = [15, 35, 35, 15]
explode = (0.2, 0, 0, 0)  # 1. slice'ı 0.2 kadar dışarıya çıkar (bu örnekte Türkiye)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```
